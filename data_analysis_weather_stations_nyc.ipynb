```python
def get_row_counts(*dfs):
    return [df.shape[0] for df in dfs]


# for faster more efficient strategy we can use map() function with getattr() method

def get_info(attr, *dfs):
    return list(map(lambda x: getattr(x, attr), dfs))
```

## Left and Right join by ranaming the matching column.

```python
"""Again if you want to practice the right and left join"""

left_join = weather_sorted.merge(station_info.rename(dict(id='station'),axis=1), on='station',how='left')
left_join
```

```python
right_join = weather_sorted.merge(station_info.rename(dict(id='station'),axis=1),on='station',how='right')
right_join
```

```python
right_join.station.describe()
```

```python
get_info('shape', left_join, right_join) 
# to get the shape of the df uing get_info(getattr method in lambda function)
```

```python
"""same joins using different method"""

left_join_using_table = station_info.merge(weather_sorted, left_on='id', right_on='station', how='left')
left_join_using_table
```

```python
right_join_using_table = weather_sorted.merge(station_info, left_on='station', right_on='id', how='right')
right_join_using_table
```

### to prove they are equivalent we need put the columns in same order, reset index and sort the data

```python
left_join_using_table.sort_index(axis=1).sort_values(['date','station']).reset_index().drop(columns='index').equals(
right_join_using_table.sort_index(axis=1).sort_values(['date','station']).reset_index().drop(columns='index'))
```

```python
get_info('shape',inner_join,left_join_using_table,right_join_using_table)
```

## Dataset consisting of duplicate, missing or invalid data

```python
messy_data_raw = pd.read_csv(r'D:\New\data_analysis_with_pandas\work_with_pandas\datasets\weather_data_with_missing_invalid_values.csv')
messy_data_raw
```

```python
messy_data = pd.read_csv('../Hands-On-Data-Analysis-with-Pandas-2nd-edition\ch_04\data\dirty_data.csv',
                        index_col='date').drop_duplicates().drop(columns='SNWD')
messy_data
```

```python
'''CREATE df for each station'''

valid_station = messy_data.query('station != "?"').copy().drop(columns=['WESF','station'])
valid_station
```

```python
station_with_wesf = messy_data.query('station == "?"').copy().drop(columns=['station','TMIN','TMAX','TOBS'])
station_with_wesf
```

```python
'''The column we want to join on (the date) is the index. so we pass the left_index to indicate the 
column we want to use from the left data frame and use the right_index from right dataframe '''

valid_station.merge(
    station_with_wesf, left_index=True, right_index=True
).query('WESF>0').head()
```

```python
valid_station.merge(
    station_with_wesf, left_index=True, right_index=True,suffixes=('','_?')
).query('WESF>0').head()
```

```python
'''Since we are joining on the index, join method is better to use here'''
'''join() method will always use the index of the left data frame, but it can also use the column
 in the right dataframe if its name is passes to "on" parameter'''
valid_station.join(station_with_wesf, rsuffix='_?',).query('WESF > 0').head()

```
